```python
from sklearn.datasets import load_wine
wine = load_wine(as_frame=True)
wine.frame.head()
```

```python
print(wine.DESCR)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, random_state=42)
```

```python
import pandas as pd
# Convert the data to a pandas DataFrame
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_df['target'] = wine.target

# Check for missing values
missing_values = wine_df.isnull().sum()

# Basic statistics of the dataset
basic_stats = wine_df.describe()

# Display the results
print("Missing Values in Wine Dataset:")
print(missing_values)
print("\nBasic Statistics of Wine Dataset:")
print(basic_stats)
#
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# Calculate the correlation matrix of the independent variables
correlation_matrix_independent = wine_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_independent, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Independent Variables in Wine Dataset')
plt.show()
```

```python
# Identify the continuous and categorical variables
continuous_variables = wine.feature_names
categorical_variables = ['target']

# Print the types of variables
print("Continuous Variables:")
print(continuous_variables)
print("\nCategorical Variables:")
print(categorical_variables)
```

```python
# Plotting the independent variables
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)

# Plot each continuous variable
plt.figure(figsize=(20, 15))
for i, column in enumerate(wine_df.columns, 1):
    plt.subplot(4, 4, i)
    sns.histplot(wine_df[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()

# Plot the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x=wine.target)
plt.title('Distribution of Target Variable')
plt.xlabel('Target Class')
plt.ylabel('Count')
plt.show()
```
